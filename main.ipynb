<h2>Setting Up Dice Rolling</h2>

```python
def dice_roll():
    dice_rolls = rd.choice([1, 2, 3, 4, 5, 6])
    return dice_rolls

def dice_cube(dice_side):
    if(dice_side == 1):
        print("""
               ____________
              |            |
              |            |
              |     ( )    |
              |            |
              |            |
               ____________
              """)
    elif(dice_side == 2):
        print("""
               ____________
              |            |
              |   ()       |
              |            |
              |            |
              |        ()  |
               ____________
              """)
    elif(dice_side == 3):
        print("""
               ____________
              |            |
              |  ()        |
              |     ()     |
              |        ()  |
              |            |
               ____________
              """)
    elif(dice_side == 4):
        print("""
               ____________
              |            |
              |   ()   ()  |
              |            |
              |   ()   ()  |
              |            |
               ____________
              """)
    elif(dice_side == 5):
        print("""
               ____________
              |            |
              |  ()    ()  |
              |     ()     |
              |  ()    ()  |
              |            |
               ____________
              """)
    elif(dice_side == 6):
        print("""
               ____________
              |            |
              |  () () ()  |
              |            |
              |  () () ()  |
              |            |
               ____________
              """)
```
